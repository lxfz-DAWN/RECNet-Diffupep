```python
import os
```

```python
def add_zdock_extension():  
        """  
        遍历指定目录下的所有文件，找到没有后缀名的文件，并给它们添加.zdock后缀。  

        :param directory: 指定目录的路径  
        """  
        # 遍历目录中的所有文件和文件夹 
        directory = './temp'
        for filename in os.listdir(directory):  
            # 构建完整的文件路径  
            file_path = os.path.join(directory, filename)
            print(file_path)
            process_list = []
            # 检查是否是文件（排除目录）  
            if os.path.isfile(file_path):  
                # 检查文件是否有后缀名（这里简单地通过是否包含'.'来判断）  
                if '.' not in filename:  
                    # 如果没有后缀名，则添加.zdock后缀  
                    new_filename = filename + '.zdock'  
                    new_file_path = os.path.join(directory, new_filename)  

                    # 重命名文件  
                    os.rename(file_path, new_file_path)  
                    print(f"Renamed '{file_path}' to '{new_file_path}'")
                    file_path_new = os.path.join(directory,new_filename)
                    try:
                        with open(file_path_new,'r') as file:
                            for line in file:
                                if line.strip() != "0.000000\t0.000000\t0.000000" :
                                    process_list.append(line)
                        with open(file_path_new,'w') as file:
                            file.writelines(process_list)
                        print('lines dealt')
                    except FileNotFoundError:
                        print('deal lines file not found')
                    except Exception as e:
                        print(f'An errro occurred : {e}')
add_zdock_extension()
                        
                
            
```
